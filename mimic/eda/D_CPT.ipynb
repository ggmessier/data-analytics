`D_CPT` table defines high-level definitions for current procedural terminology (CPT) codes.

`sectionrange` defines the range of codes for the given section, and `sectionheader` provides the description of the given section. 

There are 8 possible sections:

Evaluation and management

Surgery

Radiology

Anesthesia

Emerging technology

Pathology and laboratory

Performance measurement

Medicine

Similarly as for the sections, `subsectionrange` defines the range of codes for the given subsection, and `subsectionheader` provides the description of the given subsection. The subsection provides extra detail which can be useful, for example, when the section header is ‘Anesthesia’ the subsection provides information on the general anatomical region for the anesthesia.

`mincodeinsubsection`, `maxcodeinsubsection` provide numeric representations of minimum and maximum value in the `subsectionrange` column.
